```python
import pandas as pd
import numpy as np 
from sklearn.impute import SimpleImputer
```

```python
pharmacies_data = pd.read_csv('/content/drive/MyDrive/Домашна 1/datasets/raw_dataset.csv')
```

```python
pharmacies_data.head()
```

```python
transformation_1 = pharmacies_data.drop(pharmacies_data.columns.difference(['lat', 'lon', 'tags/name', 'tags/opening_hours', 'tags/email', 'tags/phone',	'tags/addr:street', 'tags/addr:housenumber', 'tags/website']), 1, inplace=False)
```

```python
transformation_2 = transformation_1.dropna(subset=['tags/name'], inplace=False)
```

```python
transformation_2
```

```python
transformation_3 = pd.DataFrame(transformation_2['tags/opening_hours'].fillna('08am - 17pm'))
```

```python
transformation_3.rename(columns={'tags/opening_hours':'hours'}, inplace=True)
```

```python
transformation_4 = pd.concat([transformation_2, transformation_3], axis=1)
```

```python
result = transformation_4.drop(['tags/opening_hours'],1, inplace=False)
```

```python
result
```

```python
result.to_csv('/content/drive/MyDrive/Домашна 1/datasets/filtered_dataset_final.csv')
```
